```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/task3/student_feedback.csv')  # Use correct file path if different
df.head()
```

```python
df.info()
```

```python
for col in df.columns[2:]:  # Skip ID columns
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.describe().T
```

```python
avg_scores = df.iloc[:, 2:].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_scores.plot(kind='barh', color='skyblue')
plt.title('Average Rating by Feedback Criterion')
plt.xlabel('Average Rating (110)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Overall Score'] = df.iloc[:, 2:].mean(axis=1)
df[['Student ID', 'Overall Score']].head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Overall Score'], bins=10, kde=True, color='orange')
plt.title('Distribution of Overall Feedback Scores')
plt.xlabel('Overall Score (110)')
plt.ylabel('Number of Students')
plt.grid(True)
plt.tight_layout()
plt.show()
```
